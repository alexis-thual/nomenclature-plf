```python
import re 
import pandas as pd
import urllib.request
from bs4 import BeautifulSoup
```

# Download HTML 

```python
# XLS files from https://www.performance-publique.budget.gouv.fr/documents-budgetaires/lois-projets-lois-documents-annexes-annee/exercice-2017#.WyPd0IQzZQI
# (it doesn't matter if they aren't the full nomenclatures since we only need the program numbers)

for year in [2016, 2017]:
    data = pd.read_excel("raw_data/NOM_{y}_Depenses_BG.xls".format(y=year))
    programs = data["Programme"].unique()

    for programme_id in programs: 
        print(year, programme_id)
        url = "https://www.performance-publique.budget.gouv.fr/sites/performance_publique/files/farandole/ressources/{y}/pap/html/DBGPGMJPEPGM{pid}.htm".format(pid=programme_id, y=year)

        fp = urllib.request.urlopen(url)
        res_bytes = fp.read()
        text = res_bytes.decode("utf8")
        fp.close()

        with open("downloads/{y}-{pid}.html".format(y=year, pid=programme_id), "w") as f:
            f.write(text)

```

# Test loading on one file

```python
with open("downloads/2016-175.html") as f:
    html = f.read()
    soup = BeautifulSoup(html, 'html.parser')
```

```python
actions = soup.find_all(text=re.compile("ACTION N°(.*)"));

# get the common parent div containing all the important html elements
parent = actions[0].parent.parent.parent

# Get the indexes of all the Actions
indexes = [parent.index(action.parent.parent) for action in actions]

indexes
```
