# Loading Data

````python
class Line:
    def __init__(self, row):
        self.id_tms = int(row[1])
        self.tm = row[2]
        self.id_ms = row[3]
        self.ms = row[4]
        self.id_pg = int(row[5])
        self.pg = row[6]
        self.id_a = int(row[7])
        self.a = row[8]
        # Parse sous-action if exists
        self.id_sa = int(row[9]) if len(row[9]) > 0 else None
        self.sa = row[10] if len(row[10]) > 0 else ''
````

````python
tree = Node('top', parent=None)
r = Resolver('name')

def parse_csv(file_path):
    data = []
    with open(file_path, 'r') as f:
        r = csv.reader(f, delimiter=',', quotechar='"')
        # Skip header
        next(r, None)
        # Iterate through rows
        for row in r:
            line = Line(row)
            current_tree = tree
            current_path = ['top']
            for attribute in ATTRIBUTES:
                current_path.append(getattr(line, attribute))
                try:
                    current_tree = r.get(current_tree, current_path[-1])
                except:
                    current_tree = Node(current_path[-1], parent=current_tree)

            data.append(line)

    return data
````

## Count maximal number of words

````python
def maximal_number_words(data):
    m = 0
    for line in data:
        for attribute in ATTRIBUTES:
            m = max(m, len(getattr(line, attribute).split(' ')))
    return m
````

````python
def word_count_per_attribute(data):
    l = []
    for line in data:
        for attribute in ATTRIBUTES:
            l.append(len(getattr(line, attribute).split(' ')))
    return l
````

## Load word embeddings

````python
import gensim
from gensim.models import Word2Vec
pretrained_embeddings = gensim.models.KeyedVectors.load_word2vec_format(EMBEDDING_FILE_PATH, binary=False)
````

````python
textual_data = []
datasets = [data1, data2]
for dataset in datasets:
    for line in dataset:
        for attribute in ATTRIBUTES:
            textual_data.append(getattr(line, 'a'))
````

````python
len(textual_data)
````

````python
import keras
from keras.preprocessing import sequence, text

tokenizer = text.Tokenizer(
    num_words=2e3,
    filters="!\"#$%&()*+,-./:;<=>?@[\\]^_`{|}~\t\n'´’™©®«»",
    split=" "
)

tokenizer.fit_on_texts(textual_data)
vocabulary = tokenizer.word_index.keys()
````

````python
print('Vocabulary size: {}'.format(str(len(vocabulary))))
````

````python
x = tokenizer.texts_to_sequences(["Coordination de l'action diplomatique"])
x = sequence.pad_sequences(x, PADDING)
````
